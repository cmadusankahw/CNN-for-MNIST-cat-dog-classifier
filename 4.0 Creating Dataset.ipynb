```python
import cv2
import os

categories=['Cat','Dog']   #classes (name of the folders)
catergory_dict={'Cat':[1,0],'Dog':[0,1]}
path='PetImages'
img_size=50
dataset=[]

for category in categories:
    folder_path=os.path.join(path,category)
    img_names=os.listdir(folder_path)
    
    for img in img_names:
        img_path=os.path.join(folder_path,img)
        img=cv2.imread(img_path)
        try:
            gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            resized=cv2.resize(gray,(img_size,img_size))
            dataset.append([resized,catergory_dict[category]])
        except Exception as e:
            print(e)
```

```python
import random
random.shuffle(dataset)
```

```python
data=[]
target=[]

for feature,label in dataset:
    data.append(feature)
    target.append(label)
```

```python
import numpy as np

data=np.array(data)
target=np.array(target)
data=np.resize(data,(data.shape[0],img_size,img_size,1))
```

```python
import pickle

pickle.dump(data,open('data.pickle','wb'))
pickle.dump(target,open('target.pickle','wb'))
```
