```python
import cv2

img=cv2.imread('Sample.jpg')

from matplotlib import pyplot as plt

plt.imshow(img)
```

```python
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
import numpy as np

#kernel=np.array([[1,1,1],              #smoothing 3x3
#                 [1,1,1],
#                 [1,1,1]],np.float32)/9

kernel=np.array([[0,-1,0],
                 [-1,4,-1],
                 [0,-1,0]],np.float32)   #laplacian for sharpnening

#kernel=np.ones((9,9),np.float32)/81     #smoothing 9x9

convolution=cv2.filter2D(gray,-1,kernel)
plt.imshow(convolution,cmap='gray')
```

```python
kernel=np.array([[1,2,1],              #canny horizontal edge
                 [0,0,0],
                 [-1,-2,-1]],np.float32)

convolution=cv2.filter2D(gray,-1,kernel)
plt.imshow(convolution,cmap='gray')
```

```python
kernel=np.ones((3,3),np.float32)/9
kernel=np.array([[-1,0,1],              #canny vertical edge
                 [-2,0,2],
                 [-1,0,1]],np.float32)

convolution=cv2.filter2D(gray,-1,kernel)
plt.imshow(convolution,cmap='gray')
```
